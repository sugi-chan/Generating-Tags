## build a dataset class
Pytorch does a nice job here, really just have to modify how to generate samples and how to read in data

```python
class shoe_dataset(Dataset):
    """Face Landmarks dataset."""

    def __init__(self, image_name_list, image_target_list,  root_dir, transform=None):
        """
        Args:
            csv_file (string): Path to the csv file with annotations.
            root_dir (string): Directory with all the images.
            transform (callable, optional): Optional transform to be applied
                on a sample.
        """
        self.root_dir = root_dir
        self.transform = transform
        self.image_name_list = image_name_list
        self.image_target_list = image_target_list
        

    def __len__(self):
        return len(self.image_name_list)

    def __getitem__(self, idx):
        #print(idx)
        img_name = os.path.join(self.root_dir,
                                self.image_name_list[idx])
        #print(img_name)
        image = Image.open(img_name)

        tags = np.asarray(self.image_target_list[idx])
        #print(tags)
        sample = {'image': image, 'tags': tags}
        
        if self.transform:
            sample['image'] = self.transform(sample['image'])

        return sample['image'] , sample['tags']
```

## Training Function
Standard Pytorch training function with some fun features. Basically it takes in a set of model parameters, model itself, the name to use for saving the model (it saves the best and final models) and using a dictionary of two data loaders. It runs the train loop, then uses the same loop to run the val loop. It is nicer than writing a train loop followed by a val loop. 

```python

def train_model(model, criterion, optimizer, scheduler, model_save_name,dataloaders, num_epochs=10):
    device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
    since = time.time()

    best_model_wts = copy.deepcopy(model.state_dict())
    best_loss = 500

    for epoch in range(num_epochs):
        print('Epoch {}/{}'.format(epoch, num_epochs - 1))
        print('-' * 10)

        # Each epoch has a training and validation phase
        for phase in ['train', 'val']:
            if phase == 'train':
                scheduler.step()
                model.train()  # Set model to training mode
                epoch_len = 198
         
            else:
                model.eval()   # Set model to evaluate mode
                epoch_len = 20
      
            running_loss = 0.0
            running_corrects = 0
            
            # Iterate over data.
            #for local_batch, local_labels in dataloaders[phase]:
            for local_batch, local_labels in dataloaders[phase]:

                inputs, labels = local_batch.to(device), local_labels.to(device)

                # zero the parameter gradients
                optimizer.zero_grad()

                # forward
                # track history if only in train
                with torch.set_grad_enabled(phase == 'train'):
                    outputs = model(inputs)
                    _, preds = torch.max(outputs, 1)
                    loss = criterion(outputs, labels.float())
            
                    # backward + optimize only if in training phase
                    if phase == 'train':
                        loss.backward()
                        optimizer.step()

                # statistics
                running_loss += loss.item() * inputs.size(0)

            epoch_loss = running_loss / epoch_len
            #epoch_acc = running_corrects.double() / epoch_len
            # epoch_data_loader
        
            print('{} Loss: {:.4f}'.format(
                phase, epoch_loss))#, epoch_acc))

            # deep copy the model
            if phase == 'val' and epoch_loss < best_loss:
                best_loss = epoch_loss
                torch.save(model.state_dict(), 'models/'+'best'+model_save_name)
                #best_model_wts = copy.deepcopy(model.state_dict())

        print()

    time_elapsed = time.time() - since
    print('Training complete in {:.0f}m {:.0f}s'.format(
        time_elapsed // 60, time_elapsed % 60))
    print('Best val Acc: {:4f}'.format(best_loss))

    # load best model weights
    model.load_state_dict(best_model_wts)
    return model

```

## Putting the Training Together
The main function sets the different params needed for training 

```python

import torch
import torch.nn as nn
import torch.optim as optim
from torch.optim import lr_scheduler
import numpy as np
import torchvision
from torchvision import datasets, models, transforms
import matplotlib.pyplot as plt
import time
import os
import copy
import cv2



def main():
    nodes = 5
    model_save_name = '_rest_resnet50.pth'
    device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
    '''
    currently I am just swapping through these datasets manually. Would be easy enough to just make it
    iterate over them
    '''
    # color datsets  7
    #transformed_dataset_train
    #transformed_dataset_val
    #
    #type datasets 7
    #type_dataset_train
    #type_dataset_val
    #
    # other dataset 5 
    #rest_dataset_train
    #rest_dataset_val

    #nifty data loaders! 
    trainloader=torch.utils.data.DataLoader(rest_dataset_train, batch_size=64, shuffle=True, num_workers=0)
    testloader=torch.utils.data.DataLoader(rest_dataset_val, batch_size=1, shuffle=False, num_workers=0)

    dataloaders = {}
    dataloaders['train'] = trainloader
    dataloaders['val'] = testloader


    ######################################################################
    # Finetuning the convnet
    # ----------------------
    #
    # Load a pretrained model and reset final fully connected layer.
    #
    print('building model')
    model_ft = models.resnet18(pretrained=True)
    num_ftrs = model_ft.fc.in_features
    
    model_ft.fc = nn.Linear(num_ftrs, nodes)
    model_ft = model_ft.to(device)

    criterion = nn.MultiLabelSoftMarginLoss()

    # Observe that all parameters are being optimized
    optimizer_ft = optim.SGD(model_ft.parameters(), lr=0.01,momentum=.9)

    # Decay LR by a factor of 0.1 every 7 epochs
    exp_lr_scheduler = lr_scheduler.StepLR(optimizer_ft, step_size=10, gamma=0.1)

    ######################################################################
    # Train and evaluate
    # ^^^^^^^^^^^^^^^^^^
    #
    # It should take around 15-25 min on CPU. On GPU though, it takes less than a
    # minute.
    #
    print('training model')
    model_ft = train_model(model_ft, criterion, optimizer_ft, exp_lr_scheduler,model_save_name,dataloaders,
                           num_epochs=25)
        ######################################################################
    #
    torch.save(model_ft.state_dict(), 'models/'+'final'+model_save_name)



if __name__ == '__main__':
    main()
```
